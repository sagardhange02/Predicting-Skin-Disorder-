## 1. Importing the necessary packages

```python
# Importing the necessary function
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
#Importing & defining the datframe
df=pd.read_csv('dataset_35_dermatology.csv')
df.head()
```

## 3. Checking the null Values

```python
df.replace('?',np.nan,inplace=True)
df.head(5)
```

## 4 Replacing The nulll values with mean

```python
columns = ['Age']
for col in columns:
    df[col]=pd.to_numeric(df[col])
    df[col].fillna(df[col].mean(), inplace=True)
df.head() 
```

```python
X = df.iloc[:,:34]
X.head()
```

## 5. Feature Selection with PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
x=pd.DataFrame(pca.fit_transform(X))
```

```python
x.head()
```

## 5. Defining the 'X' and 'y'

```python
x = x.iloc[:,:10]
y=df['class']
```

```python
x.head()
```

```python
y.head()
```

## 6. Model-4 => RandomForest Model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=25)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier( n_estimators=100,max_depth=15,random_state=4)
```

```python
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
```

## 7. Model Evaluation

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```

```python
confusion_matrix(y_test,y_predict)
```

## 8. Model Evaluation with Cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,X,y,cv=15,scoring="accuracy")
print(scores)
```

```python
print("Average Accuracy:",scores.mean())
print("Std:",scores.std())
#std of < 0.05 is good.
```
