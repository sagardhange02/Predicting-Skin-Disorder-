## 1. Importing the necessary packages

```python
# Importing the necessary function
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
#Importing & defining the datframe
df=pd.read_csv('dataset_35_dermatology.csv')
df.head()
```

## 2. Basic checks

```python
df.shape
```

```python
df.size
```

```python
df.info()
```

## 3. Checking the null Values

```python
df.replace('?',np.nan,inplace=True)
df.head(5)
```

```python
df.isnull().sum()
```

## 4 Replacing The nulll values with mean

```python
columns = ['Age']
for col in columns:
    df[col]=pd.to_numeric(df[col])
    df[col].fillna(df[col].mean(), inplace=True)
df.head() 
```

```python
df.isnull().sum()
```

```python
df.shape # Shape of a DataFrame
```

## 5. Defining the 'X' and 'y'

```python
X=df.iloc[:,:34]
X.head()
```

```python
y=df['class']
y.head()
```

## 6. Model-3 => RandomForest Model

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=25)
```

```python
from sklearn.ensemble import RandomForestClassifier #Defining the model
model=RandomForestClassifier(n_estimators=100,max_depth=15,random_state=4)#n_estimators=100,max_depth=15,random_state=4
```

```python
model.fit(X_train,y_train) ## fitting the data
```

```python
y_predict=model.predict(X_test)
```

## 7. Model Evaluation

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict)) # Classification Report
```

```python
confusion_matrix(y_test,y_predict) # Confusion matrix
```

## 8.Model Evaluation with Cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,X,y,cv=10,scoring="accuracy")
print(scores)
```

```python
print("Average Accuracy:",scores.mean())
print("Std:",scores.std())
#std of < 0.05 is good.
```

### 8.1 Using GridSearchCV & RandomizedSearchCV

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV 
parameters = {'max_depth':[15,16,18,19,20],
              'random_state': [1,2,3,4,5],
              'n_estimators':[80,90,100,110,120] 
             }  # dictionary of parameters.  5x5x5 = 125.
grid = GridSearchCV(model,parameters,cv=5,verbose=1)
#grid = RandomizedSearchCV(model,parameters,cv=5)
grid.fit(X,y)
```

```python
print(grid.best_score_) #Best score
print(grid.best_params_) #for best parameters
```
